```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from hparam import hps
import os
```

```python
feature_path = '../dataset/feature_augment/non-set'
```

```python
genre_dict = {}
```

```python
for file in os.listdir(feature_path):
    file_name, postfix = file.split('.')
    if postfix != 'npy':
        continue
    label = file_name.split('_')[-1]
    if label not in genre_dict.keys():
        genre_dict[label] = 1
    else:
        genre_dict[label] += 1
```

```python
keys = list(genre_dict.keys())
values = list(genre_dict.values())
```

```python
plt.figure(figsize=(12, 4))
sns.barplot(keys, values)
```

```python
shape_list = []
```

```python
for root, dirs, files in os.walk(feature_path):
    for file in files:
        file = os.path.join(root, file)
        data = np.load(file, encoding='bytes', allow_pickle=True)
        shape_list.append(data.shape[0])
```

```python
sns.stripplot(shape_list)
```
